```python
#@title CELL 3: Convert Synthea Data to GRPO Format
import pandas as pd
import json
import random
import os

# Robust path finding for Synthea output
base_path = 'synthea/output/csv'
if not os.path.exists(base_path):
    if os.path.exists('output/csv'):
        base_path = 'output/csv'
    else:
        # Fallback: try to find it relative to content if we are deep in directories
        if os.path.exists('/content/synthea/output/csv'):
            base_path = '/content/synthea/output/csv'
        else:
            raise FileNotFoundError(f"Could not find Synthea output directory. Checked {base_path} and output/csv")

print(f"Loading Synthea data from: {base_path}")

# Load Synthea outputs
procedures = pd.read_csv(f'{base_path}/procedures.csv')
conditions = pd.read_csv(f'{base_path}/conditions.csv')
claims = pd.read_csv(f'{base_path}/claims.csv')

print(f"Loaded {len(procedures)} procedures, {len(conditions)} conditions, {len(claims)} claims")

# Create GRPO training pairs with verifiable rewards
grpo_data = []

# Task 1: CPT Code Assignment (procedures)
for _, row in procedures.iterrows():
    if pd.notna(row.get('CODE')) and pd.notna(row.get('DESCRIPTION')):
        prompt = f"Patient underwent: {row['DESCRIPTION']}. What is the correct procedure code?"
        correct_answer = f"The correct code is {row['CODE']} for {row['DESCRIPTION']}."

        grpo_data.append({
            "prompt": prompt,
            "completion": correct_answer,
            "code": str(row['CODE']),
            "type": "cpt_assignment"
        })

# Task 2: ICD-10 Diagnosis Coding (conditions)
for _, row in conditions.iterrows():
    if pd.notna(row.get('CODE')) and pd.notna(row.get('DESCRIPTION')):
        prompt = f"Patient diagnosed with: {row['DESCRIPTION']}. Assign the ICD-10 code."
        correct_answer = f"ICD-10 code {row['CODE']} - {row['DESCRIPTION']}"

        grpo_data.append({
            "prompt": prompt,
            "completion": correct_answer,
            "code": str(row['CODE']),
            "type": "icd10_assignment"
        })

# Task 3: Billing Error Detection (synthetic errors)
error_types = [
    ("unbundling", "Procedures should be billed together under a single code"),
    ("upcoding", "Service billed at higher level than documented"),
    ("duplicate", "Same service billed multiple times"),
    ("medical_necessity", "Service not medically necessary per documentation")
]

for i in range(500):
    error_type, description = random.choice(error_types)
    proc = procedures.sample(1).iloc[0] if len(procedures) > 0 else None
    if proc is not None:
        prompt = f"Review this claim: {proc['DESCRIPTION']} billed as {proc['CODE']}. Check for billing errors."
        if random.random() > 0.7:  # 30% have errors
            answer = f"BILLING ERROR DETECTED: {error_type}. {description}. Recommend appeal."
            has_error = True
        else:
            answer = f"No billing errors detected. Code {proc['CODE']} appropriate for {proc['DESCRIPTION']}."
            has_error = False

        grpo_data.append({
            "prompt": prompt,
            "completion": answer,
            "has_error": has_error,
            "error_type": error_type if has_error else None,
            "type": "error_detection"
        })

print(f"Created {len(grpo_data)} training examples")

# Save to file
output_file = 'medical_billing_grpo.json'
with open(output_file, 'w') as f:
    json.dump(grpo_data, f, indent=2)

print(f"Saved training data to {output_file}")
```

```python
#@title CELL 4: Create Oumi Dataset Format
from datasets import Dataset

def format_for_oumi(examples):
    """Convert to Oumi GRPO format with reward function"""
    formatted = []
    for ex in examples:
        formatted.append({
            "prompt": ex["prompt"],
            "chosen": ex["completion"],
            "type": ex["type"]
        })
    return formatted

formatted_data = format_for_oumi(grpo_data)
dataset = Dataset.from_list(formatted_data)

# Split into train/validation
dataset = dataset.train_test_split(test_size=0.1)
print(f"Train: {len(dataset['train'])}, Validation: {len(dataset['test'])}")

# Save for Oumi
dataset.save_to_disk('medical_billing_dataset')
```

```python
#@title CELL 5: Define Reward Function for GRPO
def medical_billing_reward(prompt: str, completion: str, reference: dict) -> float:
    """
    Rule-based reward function for medical billing tasks.
    Returns reward between -1.0 and 1.0
    """
    reward = 0.0

    # Check if it's a coding task
    if "code" in reference:
        expected_code = reference["code"]
        if expected_code in completion:
            reward += 0.8  # Correct code identified
        else:
            reward -= 0.5  # Wrong code

    # Check for error detection task
    if reference.get("type") == "error_detection":
        has_error = reference.get("has_error", False)
        detected_error = "ERROR" in completion.upper() or "APPEAL" in completion.upper()

        if has_error == detected_error:
            reward += 0.5  # Correct detection
        else:
            reward -= 0.3  # Missed error or false positive

    # Format quality bonus
    if len(completion) > 20 and len(completion) < 500:
        reward += 0.2  # Good length

    return max(-1.0, min(1.0, reward))

# Test reward function
test_prompt = "Patient underwent: Appendectomy. What is the correct procedure code?"
test_completion = "The correct code is 44950 for Appendectomy."
test_ref = {"code": "44950", "type": "cpt_assignment"}
print(f"Test reward: {medical_billing_reward(test_prompt, test_completion, test_ref)}")
```

```python
#@title CELL B: Format Data for SFT Training
from datasets import load_from_disk, Dataset

# Load your saved dataset
dataset = load_from_disk('medical_billing_dataset')

# Format for SFT (instruction-response pairs)
def format_for_training(example):
    text = f"""### Instruction:
{example['prompt']}

### Response:
{example['chosen']}"""
    return {"text": text}

train_dataset = dataset['train'].map(format_for_training)
eval_dataset = dataset['test'].map(format_for_training)

print(f"Training examples: {len(train_dataset)}")
print(f"Sample:\n{train_dataset[0]['text']}")
```

```python
#@title CELL C: Train the Model with LoRA (Memory Efficient)
from transformers import AutoModelForCausalLM, AutoTokenizer
from peft import LoraConfig, get_peft_model, prepare_model_for_kbit_training
from trl import SFTTrainer, SFTConfig
from datasets import load_from_disk
import trl
import torch

print(f"Using trl version: {trl.__version__}")

# Load model with quantization (saves memory)
model_name = "Qwen/Qwen2-0.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name, trust_remote_code=True)
tokenizer.pad_token = tokenizer.eos_token

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    trust_remote_code=True,
    torch_dtype=torch.bfloat16,
    device_map="auto",
)

# LoRA config
lora_config = LoraConfig(
    r=16,
    lora_alpha=32,
    lora_dropout=0.05,
    target_modules=["q_proj", "v_proj", "k_proj", "o_proj"],
    bias="none",
    task_type="CAUSAL_LM"
)

model = get_peft_model(model, lora_config)
model.print_trainable_parameters()

# --- DATASET PREPARATION ---
# Reload and strictly format dataset to only have 'text' column
# This prevents SFTTrainer from getting confused by 'prompt'/'completion' columns
if 'dataset' not in globals():
    dataset = load_from_disk('medical_billing_dataset')

def format_clean(example):
    # Re-create text field just in case
    text = f"### Instruction:\n{example['prompt']}\n\n### Response:\n{example['chosen']}"
    return {"text": text}

# Map and REMOVE all other columns
train_dataset = dataset['train'].map(format_clean, remove_columns=dataset['train'].column_names)
eval_dataset = dataset['test'].map(format_clean, remove_columns=dataset['test'].column_names)

print(f"Dataset columns: {train_dataset.column_names}")

# --- CONFIGURATION ---
# Initialize config with standard args
training_args = SFTConfig(
    output_dir="./claimguardian_sft",
    num_train_epochs=3,
    per_device_train_batch_size=4,
    gradient_accumulation_steps=4,
    learning_rate=2e-4,
    warmup_steps=100,
    logging_steps=50,
    save_steps=500,
    eval_strategy="steps",
    eval_steps=500,
    fp16=False,
    bf16=True,
    optim="adamw_torch",
    report_to="none",
)

# Manually set TRL-specific args to avoid constructor errors
training_args.dataset_text_field = "text"
training_args.max_seq_length = 512

# --- TRAINER ---
trainer = SFTTrainer(
    model=model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    processing_class=tokenizer,
)

# START TRAINING
print("🚀 Starting training...")
trainer.train()
print("✅ Training complete!")
```
